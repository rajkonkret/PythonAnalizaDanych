watoci odstajce
IQR

```python
Q1 = df['score'].quantile(0.25)
Q3 = df['score'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
outliers = df[df['score'] > Q3 + 1.5 * IQR]
outliers
```

```python
median_score = df['score'].median()
df.loc[df['score'] > Q3 + 1.5 * IQR, 'score'] = median_score
df
```
